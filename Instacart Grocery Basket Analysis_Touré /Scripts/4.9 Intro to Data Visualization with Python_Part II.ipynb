```python
# Sampling dataframe 

np.random.seed(4)
```

```python
dev = np.random.rand(len(df_instacart_merge)) <= 0.7
```

```python
# Store 70% of the sample in the dataframe big 

big = df_instacart_merge[dev]
```

```python
# Store 30% of the sample in the dataframe small

small = df_instacart_merge[~dev]
```

```python
# Check on the numbers of rows with big and small database reunited 

len(small)+len(big)
```

```python
# Reducing your samples to only those columns necessary for the chart, namely “prices” and “order_hour_of_day”

df_2 = small[['order_hour_of_day','prices']]
```

```python
# Creating the line chart 

line = sns.lineplot(data = df_2, x = 'order_hour_of_day',y = 'prices')
```

```python
#Overall, it seems that there are no real difference. 
```

```python
# Determine whether there’s a connection between age and family situation by creating a line chart exploring the connections between age and number of dependents

df_3 = small[['family_status','age']]
```

```python
# Creating the line chart 

line2 = sns.lineplot(data = df_3, x = 'family_status',y = 'age')
```

```python
# Results line chart : People 'single' or 'liveing with parents and sibling' are more common for people from 20 to 40 years old. However, 'divorced' and 'married' are high from 50 to 70 years old.
```

```python
# Creating a scatterplot

scatterplot = sns.scatterplot(x = 'age', y = 'income',data = df_instacart_merge)
```

```python
# Scatterplot analysis : We can notice that younger customers from 20 to 40 years old has lower income comparing to customer of 50 + years old who has higher income.
```
