```python
#Aggregating Data with transform()

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Deriving Columns with loc()

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category 

loyalty_flag_stats = df.groupby('loyalty_flag').agg({'order_number': ['mean', 'min', 'max'],'prices': ['mean', 'min', 'max']})
```

```python
#Aggregating Data with transform()

ords_prods_merge['days_since_prior_order_median'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check the output 

ords_prods_merge.head(20)
```

```python
# Deriving Columns with loc()

ords_prods_merge.loc[ords_prods_merge['days_since_prior_order_median'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['days_since_prior_order_median'] > 10) & (ords_prods_merge['days_since_prior_order_median'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['days_since_prior_order_median'] <= 10, 'frequency_flag'] = 'Frequent customer'
```
